```python
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
```

```python
#finding  best random state

for i in range(1,50):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.33,random_state=i)
    l = LinearRegression()
    l.fit(x_train,y_train)
    p = l.predict(x_test)
    p1 = l.predict(x_train)
    print('r2 score of test data at random state,',i,' ',r2_score(p,y_test))
    print('r2 score of train data at random state,',i,' ',r2_score(p1,y_train))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.33,random_state=48) #using best random state to split the data
```

```python
obj_ = LinearRegression()
obj_.fit(x_train,y_train)
p=obj_.predict(x_test)
cross_val = cross_val_score(obj_,x,y,cv=4)
print("r2 score",r2_score(p,y_test))
print('cross score',cross_val.mean())
sns.regplot(x=y_test, y=p, data=df1)
```

```python
obj_ = SVR()
obj_.fit(x_train,y_train)
p=obj_.predict(x_test)
cross_val = cross_val_score(obj_,x,y,cv=4)
print("r2 score",r2_score(p,y_test))
print('cross score',cross_val.mean())
sns.regplot(x=y_test, y=p, data=df1)
```
